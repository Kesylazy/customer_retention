```python
import pandas as pd
```

```python
df_train = pd.read_csv('train_upd.csv')
df_test = pd.read_csv('test_upd.csv')
```

```python
X = df_train.drop('churn', axis=1)
y = df_train['churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
from catboost import CatBoostClassifier
```

```python
spw = len(y_train) / sum(y_train == 1)
```

```python
model = CatBoostClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
y_pred_test = model.predict(df_test)
```

```python
submission_dictionary = {
    'id': list(range(1, len(df_test) + 1)),
    'churn': y_pred_test
}
```

```python
submission_df = pd.DataFrame(submission_dictionary)
submission_df.index = list(range(1, len(df_test) + 1))
```

```python
churn_mapping = {
    0: 'no',
    1: 'yes'
}
```

```python
submission_df['churn'] = submission_df['churn'].map(churn_mapping)
```

```python
submission_df.to_csv('submission.csv', index=False)
```
